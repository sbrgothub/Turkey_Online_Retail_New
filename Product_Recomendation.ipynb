Convert product name as keys its vectors as values in a dictionary

```python
product_dictionary={}
for i,j in zip(updated_df['Product'],x_vector):
  product_dictionary[i]=[j]
```

Taking first 5 nearby vecotor products using cosine similarity

```python
def recommend(item):
  if  item in product_dictionary:
    v_in=product_dictionary[item]
    cosine_dit={}
    for prd,v_check in product_dictionary.items():
      if prd!=item:
        cosine_dit[prd]=cosine_similarity(v_in,v_check)[0][0]
    result=sorted(cosine_dit.items(),key=lambda a:a[1],reverse=True)
    return [i[0] for i in result[:5]]
  else:
    return 'not_available'
```
